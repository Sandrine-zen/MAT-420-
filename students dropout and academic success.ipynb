```python
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
```

```python
df = pd.read_csv("students dropout and academic success dataset.csv") 
df.shape
```

```python
x_col = "Age at enrollment"
y_col = "Admission grade"

df_subset = df[[x_col, y_col]].dropna()
display(df_subset.head())
```

```python
sns.jointplot(x="Age at enrollment",
              y="Admission grade",
              data=df,
              kind="reg",
              color="blue",
              height=8) 
```

```python
df2 = df.iloc[:, 6:] 
sns.heatmap(df_subset.corr(method="pearson"), cmap="coolwarm", annot=True)
plt.title("Correlation Heatmap between Age and Admission Grade")
plt.tight_layout()
plt.show() 
```
